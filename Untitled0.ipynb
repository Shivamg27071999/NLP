```python
from datetime import datetime
# Step 4: Convert the 'date' column to datetime format manually
# Remove the timezone abbreviation before parsing
def remove_timezone(date_str):
    parts = date_str.split()
    return ' '.join(parts[:-2] + parts[-1:])

df['date_clean'] = df['date'].apply(remove_timezone)

# Define a function to parse the cleaned datetime string
def parse_datetime(dt_str):
    return datetime.strptime(dt_str, '%a %b %d %H:%M:%S %Y')

# Apply the parsing function to the 'date_clean' column
df['datetime'] = df['date_clean'].apply(parse_datetime)

# Step 5: Extract time, date, and year
df['time'] = df['datetime'].dt.time
df['date_only'] = df['datetime'].dt.date
df['year'] = df['datetime'].dt.year
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
from sklearn.model_selection import train_test_split

# Assuming sampled_df is your DataFrame with processed_tweet and target columns
# Sampled DataFrame example
# sampled_df = pd.read_csv('your_processed_data.csv')

# Encode labels
label_encoder = LabelEncoder()
sampled_df['target_encoded'] = label_encoder.fit_transform(sampled_df['target'])

# Tokenize words
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sampled_df['processed_tweet'])
X_sequences = tokenizer.texts_to_sequences(sampled_df['processed_tweet'])

# Pad sequences
max_length = 300  # Maximum sequence length
X_pad = pad_sequences(X_sequences, maxlen=max_length, padding='post')

# Convert target to categorical if needed (for categorical_crossentropy loss)
# y_categorical = to_categorical(sampled_df['target_encoded'])  # Uncomment if using categorical_crossentropy

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_pad, sampled_df['target_encoded'], test_size=0.2, random_state=42)
```

```python
!pip install keras-tuner

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from keras_tuner import RandomSearch
from keras_tuner.engine.hyperparameters import HyperParameters

def build_lstm_model(hp):
    model = Sequential()

    # Embedding layer
    model.add(Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=hp.Int('embedding_dim', min_value=50, max_value=300, step=50), input_length=max_length))

    # LSTM layers
    for i in range(hp.Int('num_lstm_layers', 2, 4)):  # 2 to 3 LSTM layers
        model.add(LSTM(units=hp.Int('units_' + str(i), min_value=32, max_value=256, step=32), return_sequences=True))
        model.add(Dropout(rate=hp.Float('dropout_' + str(i), min_value=0.1, max_value=0.5, step=0.1)))

    # Final LSTM layer
    model.add(LSTM(units=hp.Int('units_final', min_value=32, max_value=256, step=32)))
    model.add(Dropout(rate=hp.Float('dropout_final', min_value=0.1, max_value=0.5, step=0.1)))

    # Output layer
    model.add(Dense(1, activation='sigmoid'))  # Sigmoid for binary classification

    # Compile model
    model.compile(optimizer='adam',
                  loss='binary_crossentropy',  # Use binary_crossentropy for sigmoid output
                  metrics=['accuracy'])

    return model

# Initialize Keras Tuner
tuner = RandomSearch(
    build_lstm_model,
    objective='val_accuracy',
    max_trials=5,  # Number of hyperparameter combinations to try
    executions_per_trial=1,  # Number of models to train per trial
    directory='lstm_tuning',
    project_name='sentiment_analysis')

# Start the search for best hyperparameter configuration
tuner.search(X_train, y_train,
             epochs=10,  # Number of epochs to train each model
             validation_data=(X_test, y_test))

# Get the best model architecture found by Keras Tuner
best_model = tuner.get_best_models(num_models=1)[0]

# Summary of the best model architecture
best_model.summary()

# Evaluate the best model on test data
loss, accuracy = best_model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy*100:.2f}%")
```
